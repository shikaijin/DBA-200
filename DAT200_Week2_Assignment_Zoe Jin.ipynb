```python
all_data = {ticker: web.get_data_yahoo(ticker,start = "2010-01-01")
            for ticker in ['JPM', 'LLY', 'GE', 'MSFT']}
price = pd.DataFrame({ticker: data['Adj Close']
                     for ticker, data in all_data.items()})
```

```python
price.boxplot(column=['JPM'])
```

```python
price.boxplot(column=['LLY'])
```

```python
price.boxplot(column=['GE'])
```

```python
price.boxplot(column=['MSFT'])
```

```python
price.describe()
```

```python
scatter_matrix(price, alpha=0.2, figsize=(10, 10))
```

```python
price.corr()
```

LLY and  MSFT stocks are the most correlated and there is a positive correlation between these two stocks. JPM and GE are the least correlated and there is a negative correlation.

```python
np.var(price)
```

MSFT has the highest variance and GE has the lowest.

```python
price.hist(column=['JPM'],bins=10)
```

The shape of JPM is nonsymmetric, bimodal.

```python
price.hist(column=['LLY'],bins=10)
```

The shape of LLY is nonsymmetric, bimodel

```python
price.hist(column=['GE'],bins=10)
```

The shape of GE is bimodal

```python
price.hist(column=['MSFT'],bins=10)
```

The shape of MSFT is right-skewed, unimodal.
