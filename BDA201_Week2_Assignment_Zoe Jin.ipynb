```python
import pandas_datareader as web
all_data = {ticker: web.get_data_yahoo(ticker,start = "2010-01-01")
            for ticker in ['JPM', 'LLY', 'GE', 'MSFT']}
price = pd.DataFrame({ticker: data['Adj Close']
                     for ticker, data in all_data.items()})
```

```python
price.boxplot(column=['JPM'])
```

```python
price.boxplot(column=['LLY'])
```

```python
price.boxplot(column=['GE'])
```

```python
price.boxplot(column=['MSFT'])
```

```python
price.describe()
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(price, alpha=0.2, figsize=(10, 10))
```

```python
price.corr()
```

LLY and  MSFT stocks are the most correlated and there is a positive correlation between these two stocks. LLY and GE are the least correlated and there is a negative correlation.
