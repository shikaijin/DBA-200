```python
import pandas as pd
import seaborn as sns
import statsmodels.formula.api as smf
import statsmodels.api as sm2
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn import metrics 
from sklearn.model_selection  import train_test_split
import numpy as np

%matplotlib inline
```

```python
df= pd.read_csv('auto.csv', header=0)
df.head()
```

```python
df.corr(method='pearson')
```

```python
# model 1
lm1 = smf.ols(formula='mpg ~  weight', data=df).fit()
lm1.summary()
```

```python
# model 2
lm_multi = smf.ols(formula='mpg ~ weight + length', data=df).fit()
lm_multi.summary()
```

The R-square=0.660 suggests model 2 fit the data reasonably well.

Model 2 has a higher R-squared (0.660) than model 1, which means that model 2 provides a better fit to the data than model that only includes weight.

Model 2 has adjusted R-square=0.631, which means not all variables are contributing.

The F statistic test the null hypothesis that all of the regression coefficients are equal to zero. For model 2, the p-val of F statistic is 4.07e-06 which is < 0.05. This means we can reject the null hypothesis, and model 2 has some validity in fitting the data.

Generally, a significance level alpha = 0.05 is used. If p val of t is < 0.05, we can reject the null hypothesis that the regression coefficient is equal to zero, and the variable affects the response. For model 2, the contribution of weight is significiant since its p-val of t = 0.040 < 0.05. The contribution of length is't significiant since its p-val of t > 0.05.
